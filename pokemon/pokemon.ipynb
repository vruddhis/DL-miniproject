```python
# Move image files into corresponding directories based on Pokemon type
for image_path in glob.glob('pokemon/*.*'):
    pokemon_name = os.path.basename(image_path).split('.')[0]
    print(pokemon_name)

    # Get pokemon type from dataframe
    match = df[df['Name'].str.lower() == pokemon_name.lower()]
    pokemon_type = match.iloc[0]['Type1']
    dir_name = pokemon_type.lower()

    type_folder = os.path.join('pokemon', dir_name)
    os.makedirs(type_folder, exist_ok=True)

    new_path = os.path.join(type_folder, f"{pokemon_name}.png")
    print("New path:", new_path)

    shutil.copyfile(image_path, new_path)
```
